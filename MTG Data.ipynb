```python
# Import dependencies
import pandas as pd # Pandas Data frames
import requests     # API Requests
```

```python
#Initial Card Data Pull
cardData = []

# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")

url = "https://api.magicthegathering.io/v1/cards"

try:
    dataRequest = requests.get(url).json()
    print ("Complete\n")
except Exception as e:
    print(dataRequest)
```

```python
len(dataRequest["cards"])
```

```python
cardData = []
for card in dataRequest["cards"] :
    cardData.append(card)
```
